```python
import pandas as pd
from googletrans import Translator
import wordninja
import re
```

```python
def is_valid(text):
    # 檢查是否只包含英文字母、數字和空格
    return bool(re.fullmatch(r"[A-Za-z0-9\s]*", text))
```

```python
def translate_hashtag(text):
    text = text[3:-1]
    buff = text.split("\', \'#")
    print(buff)
    trans_buff = []
    have_error = False
    is_empty = False
    if buff[0] != '':
        for i in buff:
            if i is not None:
                try:
                    translated = translator.translate(i, src='auto', dest='en')
                    trans_buff.append(translated.text)
                except Exception as e:
                    try:
                        de_lang = translator.detect(i)
                        if de_lang == "en":
                            tmp = " ".join(wordninja.split(i))
                            trans_buff.append(tmp)
                        elif is_valid(i):
                            trans_buff.append(i)
                        else:
                            have_error = True
                            continue
                    except Exception as e:
                        have_error = True
                        continue
        print(trans_buff)
    else:
        is_empty = True

    if len(trans_buff) == 0 or (len(trans_buff) == 1 and trans_buff[0] == ''):
        is_empty = True
        return "no hashtage", have_error, is_empty
    else:
        return " ,".join(trans_buff), have_error, is_empty
```

```python
# 讀取 CSV 文件，跳過錯誤的行
df = pd.read_csv('/content/output.csv', on_bad_lines='skip')

# 假設你要翻譯名為 'hashtag' 和 'Pic Description' 的列
text1 = df['hashtag'].tolist()
text2 = df['Pic Description'].tolist()

# 初始化翻譯器
translator = Translator()
```

```python
trans_hash_ls = []
err_ls = []
is_empty_ls = []
for i in text1:
    trans_hash, err, is_empty = translate_hashtag(i)
    trans_hash_ls.append(trans_hash)
    err_ls.append(err)
    is_empty_ls.append(is_empty)

df['hashtag'] = trans_hash_ls
df['have_error'] = err_ls
df['is_empty'] = is_empty_ls

output_path = '/content/translated_output.csv'
df.to_csv(output_path, index=False)

print(f"翻譯完成，已保存到 {output_path}")
```

```python
def translate_Pic(text):
    trans_buff = "No Pic Description"
    have_error = False
    is_empty = (text == "nan" or text == '')
    if is_empty:
        return trans_buff, have_error, is_empty

    try:
        translated = translator.translate(text, src='auto', dest='en')
        trans_buff = translated.text
    except Exception as e:
        have_error = True

    return trans_buff, have_error, is_empty
```

```python
# 讀取 CSV 文件，跳過錯誤的行
df = pd.read_csv('/content/translated_output.csv', on_bad_lines='skip')

# 假設你要翻譯名為 'hashtag' 和 'Pic Description' 的列
text2 = df['Pic Description'].tolist()

# 初始化翻譯器
translator = Translator()
```

```python
trans_pic_ls = []
err_ls = []
is_empty_ls = []
for i in text2:
    trans_pic, err, is_empty = translate_Pic(str(i))
    trans_pic_ls.append(trans_pic)
    err_ls.append(err)
    is_empty_ls.append(is_empty)
    print(str(i), trans_pic)

df['Pic Description'] = trans_pic_ls
df['Pic_have_error'] = err_ls
df['Pic_is_empty'] = is_empty_ls

output_path = '/content/translated_output2.csv'
df.to_csv(output_path, index=False)

print(f"翻譯完成，已保存到 {output_path}")
```
